```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1번 데이터 읽어오기

df = pd.read_csv('NFLX.csv', encoding='cp949', usecols=['Date', 'Open', 'High', 'Low', 'Close'])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
# 2번 2021 이후 종가 데이터 출력하기

f_df = df.query("Date >= '2021-01-01'")
f_df.head(5)
```

```python
# 2번 2021 이후 종가 데이터 시각화하기

x = f_df['Date']  # x 좌표값
y = f_df['Close']  # y 좌표값

# 그래프 그리기
plt.plot(x, y)

# 그래프에 제목과 축 레이블 추가
plt.title('close data')
plt.xlabel('Date')
plt.ylabel('close')

# 그래프 표시
plt.show()
```

```python
# 3번 2021 이후 최고, 최저 종가 구하기

max_price = f_df['Close'].max()
min_price = f_df['Close'].min()
print('최고 종가:', max_price)
print('최고 종가:', min_price)
```

```python
# 4번 2021년 이후 월별 종가 평균 구하기 + 시각화

f_df['datetime_m'] = f_df['Date'].dt.strftime('%Y-%m')
f_g_df = f_df.groupby('datetime_m').mean()

f_g_df

x = f_g_df['Date']  # x 좌표값
y = f_g_df['Close']  # y 좌표값

# 그래프 그리기
plt.plot(x, y)

# 그래프에 제목과 축 레이블 추가
plt.title('close date_m_g')
plt.xlabel('Date_m')
plt.ylabel('close')

# 그래프 표시
plt.show()
```

```python
# 5번 2022 이후 최고, 최저, 종가 시각화하기

f_df_2 = df.query("Date >= '2022-01-01'")
# f_df_2.head()


x = f_df_2['Date']  # x 좌표값
y = f_df_2['High']  # y2 좌표값
y2 = f_df_2['Low']
y3 = f_df_2['Close']


# 그래프 그리기
plt.plot(x, y, label='High')
plt.plot(x, y2, label='Low')
plt.plot(x, y3, label='Close')

# 축 기울이기
plt.xticks(rotation=45)

# 범례 추가
plt.legend()

# 그래프에 제목과 축 레이블 추가
plt.title('Simple Line Plot')
plt.xlabel('date')
plt.ylabel('price')

# 그래프 표시
plt.show()
```
